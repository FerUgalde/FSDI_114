```python
class QueueII:
    class __Node:
        def __init__(self, data):
            self.data = data
            self.next = None

    def __init__(self):
        self.front = None
        self.last = None
        self.length = 0

    def enqueue(self, value):
        # create a new node 
        # if the queue is empty the self.front and self.last is equals to the new node
        # else, self.last.next and self.last is equals to the new node 
        # increase 1 the sizce of the queue
        new_node = self.__Node(value)
        if self.length == 0:
            self.front = new_node
            self.last = new_node
        else:
            self.last.next = new_node
            self.last = new_node
        self.length += 1

    def dequeue(self):
        # if self.front and self.lasr are equals, self.front and self.last are equals to None
        # else save self.front in a variable temp
        # now self.front is equals to temp.next
        # temp.next now is equals None
        # decrease 1 the sizce of the queue
        if self.front == self.last:
            datum = self.front.data
            self.front = None
            self.last = None
            return datum
        else:
            temp = self.front
            self.front = temp.next
            temp.next = None
        self.length -= 1

    # nice to have methods
    def size(self):
        # return the self.length
        return self.length

    def peek(self):
        return self.front.data

    def list(self):
        # save the self.front in temp (temporal variable)
        # while temp is not equals to None print the temp.data
        # the temp update with the next of the node
        temp = self.front
        while temp is not None:
            print(temp.data)
            temp = temp.next

    def is_empty(self):
        # the self.front and self.last will be equals to None
        # the size of the node will be equals to 0
        self.front = None
        self.last = None
        self.length = 0
```

```python
class SinglyLinkedList:
    class __Node:
        def __init__(self, data):
            self.data = data
            self.next = None

    def __init__(self):
        self.head = None
        # Challenge for you: see if you can have and mantain a tail ref and how that impacts performance
        self.length = 0

    def append(self, value):
        new_node = self.__Node(value)
        if not self.head:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node
        self.length += 1

    def prepend(self, value):
        new_node = self.__Node(value)
        if not self.head:
            self.head = new_node
        else:
            current = self.head
            self.head = new_node
            self.head.next = current
        self.length += 1

    def insert(self, index, value):
        # This method has to insert a new node (with value) BEFORE the index, not after:
        if index > self.length:
            print("Your position exceeds the size of the list.") 
        elif index == 0:
            self.prepend(value)
        elif index == self.length:
            self.append(value)
        else:
            new_node = self.__Node(value)
            current = self.head
            for x in range(index-1):
                current = current.next
            new_node.next = current.next
            current.next = new_node
            self.length += 1
            

    def __str__(self):
        # This dunder method must return a string at all times
        out = "["
        current = self.head
        if current:
            out += "%s" % repr(current.data)
            current = current.next
            while current:
                out += ", %s" % repr(current.data)
                current = current.next
        out += "]"
        return out
```
