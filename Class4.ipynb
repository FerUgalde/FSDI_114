```python
# Doubly Linked List

class DoublyLinkedList:
    class __Node:
        def __init__(self, data):
            self.data = data
            self.next = None
            self.prev = None

    def __init__(self):
        self.head = None
        self.tail = None
        self.size = 0

    def append(self, value):
        new_node=self.__Node(value)
        if self.head is None:
            self.head=new_node
            self.tail=self.head
        else:
            self.tail.next=new_node
            new_node.prev=self.tail
            self.tail=new_node
        self.size=1

    def insert(self, index, value):
        new_node = self.__Node(value)
        if index <= 0 :
            self.head.prev = new_node
            new_node.next = self.head
            self.head = new_node
        elif index > self.size:
            self.append(value)
        else:
            current = self.head
            for _ in range (index-1):
                current = current.next
            new_node.next = current.next
            current.next.prev = new_node
            current.next = new_node
            new_node.prev = current
        self.size += 1
        return new_node.data
    
    def remove(self, value):
        # Remove the first instance
        current = self.head
        prev = None
        found = False
        while current and not found:
            if current.data == value:
                found = True
            else:
                prev = current
                current = current.next
        if found:
            if not prev:
                self.head = self.head.next
            else:
                if current == self.tail:
                    self.tail = self.tail.prev
                next_node = current.next
                prev.next = next_node
                if next_node:
                    next_node.prev = prev
            self.size -= 1
        else:
            raise ValueError("Value not found")

    def index(self, value):
        # Return first index of value
        current = self.head
        count = 0
        found = False
        while current.next and not found:
            if current.data == value:
                found = True
            else:
                current = current.next
                count += 1
        if found:
            return count
        raise ValueError("Value not found")
    
    def search(self, index):
        current = self.head
        count = 0
        found = False
        while current.next and not found:
            if count == index:
                found = True
            else:
                current = current.next
                count += 1
        if found:
            return current.data
        raise ValueError("Index out of bounds")

    def __str__(self):
        # This dunder method must return a string at all times
        out = "["
        current = self.head
        if current:
            out += "%s" % repr(current.data)
            current = current.next
        while current:
            out += ", %s" % repr(current.data)
            current = current.next
        out += "]"
        return out

    def is_empty(self):
        return self.size == 0

    def __getitems__(self, index):
        return self.search(index)

    def __setitems__(self, index, value):
        change_node = self.head
        if index < 0 or index >= self.size:
            raise IndexError("Index out of bounds")
        for _ in range(index):
            change_node = change_node.next
        change_node.data = value
        print(change_node.data)
        
    
```

```python
class Matrix:
    def __init__(self, rows, cols):
        self.rows = rows
        self.cols = cols
        self.create()

    def create(self):
        matrix = DoublyLinkedList()
        for _ in range(self.rows):
            row = DoublyLinkedList()
            for _ in range(self.cols):
                row.append(0)
            matrix.append(row)
        self.matrix = matrix

    def __str__(self):
        matrix_str = ''
        current_row = self.matrix.head
        while current_row:
            current_col = current_row.data.head
            while current_col:
                matrix_str += str(current_col.data) + ' '
                current_col = current_col.next
            matrix_str += '\n'
            current_row = current_row.next
        return matrix_str
```
